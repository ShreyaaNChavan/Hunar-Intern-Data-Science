## Problem 2: SQL Queries – Student & Course Database
### Objective

- Create Student and Course tables and perform required SQL queries.

```python
import sqlite3

# Create database (or connect if it already exists)
conn = sqlite3.connect("student_course.db")

# Create cursor object
cursor = conn.cursor()

print("Database connected successfully")
```

```python
# Create Student table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    RollNo INTEGER PRIMARY KEY,
    Name TEXT,
    Marks INTEGER,
    AadharNo INTEGER UNIQUE,
    Address TEXT
)
""")

# Create Course table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Course (
    RollNo INTEGER,
    Course TEXT,
    Course_Duration TEXT,
    FOREIGN KEY (RollNo) REFERENCES Student(RollNo)
)
""")

conn.commit()
print("Tables created successfully")
```

```python
# Insert data into Student table
student_data = [
    (1, 'Rahul', 85, 123456789012, 'Mumbai'),
    (2, 'Riya', 25, 234567890123, 'Pune'),
    (3, 'Amit', 60, 345678901234, 'Delhi'),
    (4, 'Rohan', 20, 456789012345, 'Nashik')
]

cursor.executemany(
    "INSERT OR IGNORE INTO Student VALUES (?, ?, ?, ?, ?)",
    student_data
)

# Insert data into Course table
course_data = [
    (1, 'BCA', '3 Years'),
    (2, 'BSc IT', '3 Years'),
    (3, 'BCA', '3 Years'),
    (4, 'BCom', '3 Years')
]

cursor.executemany(
    "INSERT INTO Course VALUES (?, ?, ?)",
    course_data
)

conn.commit()
print("Data inserted successfully")
```

```python
cursor.execute("SELECT AVG(Marks) FROM Student")
avg_marks = cursor.fetchone()

print("Average Marks:", avg_marks[0])
```

```python
cursor.execute("SELECT Name FROM Student ORDER BY Name ASC")
names = cursor.fetchall()

print("Names in Ascending Order:")
for name in names:
    print(name[0])
```

```python
cursor.execute("SELECT RollNo, Name FROM Student WHERE Marks < 30")
low_score_students = cursor.fetchall()

print("Students scoring below 30:")
for student in low_score_students:
    print(student)
```

```python
cursor.execute("SELECT RollNo FROM Student WHERE Name LIKE 'R%'")
roll_numbers = cursor.fetchall()

print("Roll Numbers starting with 'R':")
for roll in roll_numbers:
    print(roll[0])
```

```python
cursor.execute("SELECT RollNo FROM Course WHERE Course = 'BCA'")
bca_students = cursor.fetchall()

print("Roll Numbers of BCA students:")
for roll in bca_students:
    print(roll[0])
```

```python
conn.close()
print("Database connection closed")
```
